```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv("Iris.csv")
```

```python
sb.lmplot(x='SepalLengthCm', y='SepalWidthCm', data=df, hue='Species', fit_reg=True)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width')
plt.show()
```

```python
sb.lmplot(x='PetalLengthCm', y='PetalWidthCm', data=df, hue='Species', fit_reg=True)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs. Petal Width')
plt.show()
```

```python
# Box plot for Sepal Length by Species
sb.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length by Species')
plt.show()
```

```python
# Box plot for Sepal Width by Species
sb.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Width by Species')
plt.show()
```

```python
# Box plot for Petal Length by Species
sb.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Petal Length by Species')
plt.show()
```

```python
# Box plot for Petal Width by Species
sb.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Width by Species')
plt.show()
```

```python
# Histogram for Sepal Length
plt.hist(df['SepalLengthCm'], bins=20, edgecolor='White', color='HotPink')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# Histogram for Sepal Width
plt.hist(df['SepalWidthCm'], bins=20, edgecolor='White', color='HotPink')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python
# Histogram for Petal Length
plt.hist(df['PetalLengthCm'], bins=20, edgecolor='White', color='HotPink')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

```python
# Histogram for Petal Width
plt.hist(df['PetalWidthCm'], bins=20, edgecolor='White', color='HotPink')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.show()
```

```python
# Joint plot for Sepal Length vs. Sepal Width
sb.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df, kind='hex', color='purple')
plt.show()
```

```python
# Joint plot for Petal Length vs. Petal Width
sb.jointplot(x='PetalLengthCm', y='PetalWidthCm', data=df, kind='hex', color='purple')
plt.show()
```

```python
# Horizontal bar plot for top 10 Species by Sepal Length
name = df['Species'][:10]
sepal_length = df['SepalLengthCm'][:10]
colors = ['red', 'yellow', 'green', 'blue', 'gold', 'orange', 'white', 'black', 'cyan', 'brown']
plt.barh(name, sepal_length, linewidth=0.7, edgecolor='red', color=colors)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Species')
plt.title('Top 10 Species by Sepal Length')
plt.show()
```

```python
# Pie chart for Species distribution
plt.pie(df['Species'].value_counts(), labels=df['Species'].unique(), shadow=True, startangle=90, autopct='%1.2f%%')
plt.title("Species Distribution")
plt.axis('equal')
plt.show()
```

```python
# Line plot for Sepal Length
plt.plot(df['SepalLengthCm'], color='green')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length')
plt.show()
```

```python
# Scatter plot for Sepal Length vs. Petal Length
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], color='Gold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title("Sepal Length vs. Petal Length")
plt.show()
```

```python
# Scatter plot for Sepal Width vs. Petal Width
plt.scatter(df['SepalWidthCm'], df['PetalWidthCm'], color='Gold')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title("Sepal Width vs. Petal Width")
plt.show()

# DataFrame describe
print(df.describe())

# DataFrame info
print(df.info())
```
